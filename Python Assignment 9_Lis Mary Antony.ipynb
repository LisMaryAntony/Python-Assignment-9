# Exercise 4

```python
data = {
    'name': ['Annamma', 'Baby', 'Charles', 'David', 'Elizabeth', 'Ferin', 'Gemma', 'Hima', 'Ivan', 'Jose'],
    'age': [28, 34, 23, 45, 30, 29, 40, 35, 22, 31],
    'gender': ['F', 'M', 'M', 'M', 'F', 'F', 'F', 'F', 'M', 'M']
}
```

```python
df = pd.DataFrame(data)
print("Data framework:\n", df)
```

# Exercise 5

```python
occupations = ['Programmer', 'Librarian', 'Analyst', 'Manager', 'Nurse', 'Tester', 'Nurse', 'Lecturer', 'Analyst', 'Teacher']
df['occupation'] = occupations
print("Updated Data framework:\n", df)
```

# Exercise 6

```python
age_filtered_df = df[df['age'] >= 30]
print("Age>= 30: \n", age_filtered_df)
```

# Exercise 8

```python
import matplotlib.pyplot as plt
```

```python
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016]
city_a = [500000, 550000, 600000, 650000, 700000, 750000, 800000]
city_b = [800000, 850000, 900000, 950000, 1000000, 1050000, 1100000]
city_c = [1000000, 1050000, 1100000, 1150000, 1200000, 1250000, 1300000]
city_d = [1200000, 1250000, 1300000, 1350000, 1400000, 1450000, 1500000]
```

```python
plt.plot(years, city_a, label='City A')
plt.plot(years, city_b, label='City B')
plt.plot(years, city_c, label='City C')
plt.plot(years, city_d, label='City D')

plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population of Cities Over Time')
plt.legend()
plt.show()
```
